# shapeShift

*sarpy*.shapeShift(img, c)

Shift image by a certain factor.
    
Parameters:

* img: ndarray
  - Input binary image
    
    
* c: {float, tuple of floats}
   - Scale factors. Separate factors can be defined as (row_scale, col_scale)
    

Returns:

* g: ndarray
    - Shifted version of the input image
