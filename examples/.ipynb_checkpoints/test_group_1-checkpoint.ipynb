# Shape scale
- sarpy.binaryScale.shapeScale

```python
def visualize(data,nclass):
    plt.rcParams["figure.figsize"] = [10,3]    
    x = 5;
    y = nclass
    fig, axes = plt.subplots(y,x)
    k = i = j = 0
    idx = 0;
    for char in data:
        idx = 0
        for sc in [4,2,1,0.8,0.5]:
            axes[i,idx].axis('off')
            axes[i,idx].imshow(sarpy.binaryScale.shapeScale(char,sc,(char.shape[0]/2,char.shape[1]/2)),cmap='gray')
            idx+=1
        i = i + 1        
        
    plt.show()

visualize(dataset['bitmaps'][::6000],len(dataset['bitmaps'][::6000]))
```

```python
aContour = np.array([[0,2,2],[1,1,3],[2,1,4],[3,1,5],
                     [4,2,6],[5,3,5],[6,4,5],[7,5,6],
                     [8,6,5],[9,6,4],[10,6,3],[11,5,2],
                     [12,4,1],[13,3,1],[14,2,2]]);
import numpy as np

from skimage import measure

from sarpy import extractContours
from sarpy import plotContour1D
plt.rcParams["figure.figsize"] = [15,6]   

def extract_contours(img):
    contours = measure.find_contours(img, 0)
    contours_lens = np.array([len(c) for c in contours])
    sort_order = (-contours_lens).argsort()
    return([[np.array([j, contours[i][j][0], contours[i][j][1]], dtype = int) for j in range(len(contours[i]))] for i in sort_order])
  
    
def fun(c,l=100):
    a = np.array([0,0,0])
    for i in c[0][:]:
        a = np.vstack((a,i))
    a = a[1::]
    
    b = np.array(a[0])
    s = a.shape[0]
    for i in range(1,l):
        idx = int(np.floor(s*i/l))
        b = np.vstack((b,a[idx]))
    return b

numbins = 200
cc = []
fig = plt.figure()
ax = fig.add_subplot(111)
qq = np.arange(0,numbins)

for i in range(103,126):
    aContour = extract_contours(dataset['bitmaps'][i])    
    img = fun(aContour,numbins)
    cc.append(img)
#     ax.plot(qq, img[:,1])

dd = np.array(cc)
mm = np.mean(dd,axis=0)
std = np.std(dd,axis=0)
ax.plot(qq, mm[:,1],'+', color="black")
plt.errorbar(qq,mm[:,1],yerr=std[:,1],ecolor='yellow',elinewidth=3,zorder=-1,alpha=1)
# plt.errorbar(qq,mm[:,1],yerr=std[:,1],ecolor='red',elinewidth=1,zorder=10,alpha=1)

plt.show()
```

```python
def bitmap2setPoints(bitmap,adj=4):
    adjx = []
    adjy = []
    if adj == 8:
        adjx = [-1,0,1,-1,1,-1,0,1];
        adjy = [-1,-1,-1,0,0,1,1,1];
    else:
        adjx = [0,-1,1,0];
        adjy = [-1,0,0,1];
        
    adjn = len(adjx)
    nrows,ncols = bitmap.shape
    setPoints = set();
    for i in range(0,nrows):
        for j in range(0,ncols):
            if bitmap[i,j] == 1:
                for dx,dy in zip(adjx,adjy):        
                    x = i + dx
                    y = j + dy
                    if x >= 0 and x < ncols and y >=0 and y < nrows:
                        if bitmap[x,y] == 0:                        
                            setPoints.add(tuple([x,y]))
    return np.array(list(setPoints))

# def normalize (shape):
    
img = dataset['bitmaps'][1777]
# plt.imshow(img)
# plt.show()
a = bitmap2setPoints(img,8)
print(type(a))
min_i,min_j = a.min(axis=0)
max_i,max_j = a.max(axis=0)
newa = a - [min_i,min_j]
img_out = np.zeros((max_i-min_i+1,max_j-min_j+1))#     print(p)
for p in newa:
    img_out[p[0],p[1]] = 1
plt.imshow(img_out)
plt.show()
```
